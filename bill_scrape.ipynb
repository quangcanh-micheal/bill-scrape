```python
import pandas as pd
from serpapi import GoogleSearch
```

```python
# 1. Setup Configuration
API_KEY = "api_key"  # Replace with your actual SerpApi key

# 2. Read the Excel file
df = pd.read_excel("receipt_query.xlsx")

# Initialize a list to store results
all_image_links = []
```

```python
# 3. Iterate through the queries
for query in df['query_tripadvisor']:
    print(f"Searching for: {query}")
    
    params = {
        "q": query,
        "tbm": "isch",         # Search type: Images
        "engine": "google",
        "api_key": API_KEY
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    
    # Extract the first 5 image URLs
    images = results.get("images_results", [])
    links = [img.get("original") for img in images[:5]]
    
    all_image_links.append(links)
```

```python
# 4. Add results back to dataframe and save
# We'll join the links into a single string or create new columns
df['top_5_image_links'] = [", ".join(l) for l in all_image_links]
df.to_excel(r"C:\Users\DELL\Desktop\receipt_results.xlsx", index=False)

print("Scraping complete! Results saved to receipt_results.xlsx")
```
